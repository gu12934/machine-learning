Feature engineering:
Aggregate

```python
airbnb1.groupby(['neighbourhood_group']).count()
airbnb1.groupby(['neighbourhood_group']).count()
airbnb1[['neighbourhood_group', 'price']]
```

```python
from collections import defaultdict
neighbourhood_price_dic = defaultdict(list)
```

```python
np_zip = zip(airbnb1.neighbourhood_group, airbnb1.price)
for i in np_zip:
 neighbourhood_price_dic[i[0]].append(i[1])
```

```python
import numpy as np
price_dic=defaultdict(list)
price_data = []
for k, price in neighbourhood_price_dic.items():
	price_data= [max(price),min(price), np.median(price), np.mean(price)]
	price_dic[k].append(price_data)
```

```python
airbnb1_neighbourhood= pd.DataFrame.from_dict(price_dic,orient='index',columns = ['data'])
```

```python
airbnb1_neighbourhood["max_price"]=airbnb1_neighbourhood["data"].apply(lambda x: x[0])
airbnb1_neighbourhood["min_price"]=airbnb1_neighbourhood["data"].apply(lambda x: x[1])
airbnb1_neighbourhood["median_price"]=airbnb1_neighbourhood["data"].apply(lambda x: x[2])
airbnb1_neighbourhood["mean_price"]=airbnb1_neighbourhood["data"].apply(lambda x: x[3])
airbnb1_neighbourhood.drop(columns=["data"],inplace=True)
display (airbnb1_neighbourhood)
```
